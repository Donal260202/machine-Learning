```python
##Activation Function
```

```python
#RELU:Rectified linear unit(ELEMENT below 0 are considered 0)
#f(x)=x if x>=0
#f(x)=0 if x<0
```

```python
def relu(x):
    return max(0,x)

    
```

```python
relu(-6)
```

```python
xl=np.linspace(-10,10)
y=list(map(lambda x: max(0,x),xl))
```

```python
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.title("ReLU FUNCTION")
plt.xlabel("XL")
plt.ylabel("RELU(XL)")
plt.show()
```

```python
#LEAKY RELU
```

```python
xl=np.linspace(-10,10)
y=list(map(lambda x: max(0.1*x,x),xl))
```

```python
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.title(" LEAKY ReLU FUNCTION")
plt.xlabel("XL")
plt.ylabel("Leaky RELU(XL)")
plt.show()
```

```python
#SIGMOID Activation Function
```

```python

xl=np.linspace(-10,10)
y=list(map(lambda x: 1/(1+np.exp(-x)),xl))
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.title(" Sigmoid FUNCTION")
plt.xlabel("XL")
plt.ylabel("Sigmoid(XL)")
plt.show()
```

```python
#HYPERBOLIC TANGENT
xl=np.linspace(-10,10)
y=list(map(lambda x: np.tanh(x),xl))
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.title(" Tanh FUNCTION")
plt.xlabel("XL")
plt.ylabel("Tanh(XL)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):

    exp_x = np.exp(x - np.max(x))  
    return exp_x / np.sum(exp_x)


logits = np.linspace(-10, 10, 100)


softmax_values = softmax(logits)


plt.figure(figsize=(8, 6))
plt.plot(logits, softmax_values)
plt.title("Softmax Activation Function")
plt.xlabel("Logits")
plt.ylabel("Softmax Probability")

plt.show()
```
